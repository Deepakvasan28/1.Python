```python
#9.What will happen when you run the following code?


count = 10

def increment():
    count += 1
    print(count)

increment()
```

```python
#ans - UnboundLocalError: local variable 'count' referenced before assignment
#count is a global variable
#Inside the function, Python sees count += 1
#This makes count a local variable inside increment()
#But you try to use it before assigning a local value
#Hence, Python throws an error

count = 10

def increment():
    global count
    count += 1
    print(count)

increment()
```
